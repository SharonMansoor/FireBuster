```python
from pydap.client import open_url
from pydap.cas.urs import setup_session
import datetime as dt  # Python standard library datetime  module
import numpy as np
from netCDF4 import Dataset  # http://code.google.com/p/netcdf4-python/
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap, addcyclic, shiftgrid
import pandas as pd
from datetime import datetime
from datetime import timedelta

# Global Vars
BASE_URL = 'E:/DataForDS/NCALDAS_NOAH0125_D.A'
END_URL = '.002.txt'
```

```python
#Read Fire data
data = pd.read_csv("C:/Fires/Top1000FireLocationsCases-21.csv")
```

```python
myDateFormat = '%m/%d/%Y'
lats = np.genfromtxt('NoahLats.csv', delimiter=',')
lons = np.genfromtxt('NoahLons.csv', delimiter=',')
lats
```

```python
#Dis = map(lambda x: datetime.strptime(x,myDateFormat),data['DISCOVERY_DATE'])
import gc

lats = np.genfromtxt('NoahLats.csv', delimiter=',')
lons = np.genfromtxt('NoahLons.csv', delimiter=',')
import timeit

start = timeit.default_timer()

i=0
oldDate = None
dataset = None
Snowf=[]
Rainf=[]
RadT=[]
SnowFrac=[]
Tair_f=[]
Tair_f_min=[]    
Tair_f_max=[]
SoilMoist0_10cm=[]
SoilMoist10_40cm =[]
SoilTemp0_10cm=[]
SoilTemp10_40cm=[]
Wind_f=[]
Qair_f=[]
Psurf_f=[]
    
for indexes,fire in data.iterrows():
    date = datetime.strptime(str(fire['DISCOVERY_DATE']),myDateFormat)
    year = str(date.year)
    month = '{:02d}'.format(date.month)
    day = '{:02d}'.format(date.day)
    FullURL = BASE_URL+year+month+day+END_URL
    
    if (year != '2016'):
        dataset = Dataset(FullURL, 'r')
        lat_idx = fire['X_tile']
        lon_idx = fire['Y_tile']
    
        Snowf.append(dataset.variables['Snowf'][:][0,lat_idx,lon_idx].data.tolist())
        Rainf.append(dataset.variables['Rainf'][:][0,lat_idx,lon_idx].data.tolist())
        RadT.append(dataset.variables['RadT'][:][0,lat_idx,lon_idx].data.tolist())
        SnowFrac.append(dataset.variables['SnowFrac'][:][0,lat_idx,lon_idx].data.tolist())
        Tair_f.append(dataset.variables['Tair_f'][:][0,lat_idx,lon_idx].data.tolist())
        Tair_f_min.append(dataset.variables['Tair_f_min'][:][0,lat_idx,lon_idx].data.tolist())   
        Tair_f_max.append(dataset.variables['Tair_f_max'][:][0,lat_idx,lon_idx].data.tolist())
        SoilMoist0_10cm.append(dataset.variables['SoilMoist0_10cm'][:][0,lat_idx,lon_idx].data.tolist())
        SoilMoist10_40cm.append(dataset.variables['SoilMoist10_40cm'][:][0,lat_idx,lon_idx].data.tolist())
        SoilTemp0_10cm.append(dataset.variables['SoilTemp0_10cm'][:][0,lat_idx,lon_idx].data.tolist())
        SoilTemp10_40cm.append(dataset.variables['SoilTemp10_40cm'][:][0,lat_idx,lon_idx].data.tolist())
        Wind_f.append(dataset.variables['Wind_f'][:][0,lat_idx,lon_idx].data.tolist())
        Qair_f.append(dataset.variables['Qair_f'][:][0,lat_idx,lon_idx].data.tolist())
        Psurf_f.append(dataset.variables['Psurf_f'][:][0,lat_idx,lon_idx].data.tolist())
        dataset.close()
    else:
        Snowf.append(np.NaN)
        Rainf.append(np.NaN)
        RadT.append(np.NaN)
        SnowFrac.append(np.NaN)
        Tair_f.append(np.NaN)
        Tair_f_min.append(np.NaN)
        Tair_f_max.append(np.NaN)
        SoilMoist0_10cm.append(np.NaN)
        SoilMoist10_40cm.append(np.NaN)
        SoilTemp0_10cm.append(np.NaN)
        SoilTemp10_40cm.append(np.NaN)
        Wind_f.append(np.NaN)
        Qair_f.append(np.NaN)
        Psurf_f.append(np.NaN)
    
    i=i+1
    if (i%1000 == 0):
        stop = timeit.default_timer()
        print('Time: ', stop - start , (i/20000)*100,"% is Done")
        start = timeit.default_timer()

data['Snowf']=Snowf
data['Rainf']=Rainf
data['RadT']=RadT
data['SnowFrac']=SnowFrac
data['Tair_f']=Tair_f
data['Tair_f_min']=Tair_f_min
data['Tair_f_max']=Tair_f_max
data['SoilMoist0_10cm']=SoilMoist0_10cm
data['SoilMoist10_40cm']=SoilMoist10_40cm
data['SoilTemp0_10cm']=SoilTemp0_10cm
data['SoilTemp10_40cm']=SoilTemp10_40cm
data['Wind_f']=Wind_f
data['Qair_f']=Qair_f
data['Psurf_f']=Psurf_f
        
data.to_csv('C:/Fires/Top1000LoactionsWithNoahData-21.csv',mode ='a',index=False)
```
